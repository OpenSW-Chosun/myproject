```python
import matplotlib.pyplot as plt
import numpy as np

# 예시 데이터
x = np.linspace(0, 10, 100)
y = np.sin(x)

# 그래프 생성
plt.figure(figsize=(8, 6))  # 그래프 크기 설정
plt.plot(x, y, label="Sine Wave", color='b')

# 그래프 제목 및 레이블 설정
plt.title("Sine Wave")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")

# 범례 추가
plt.legend()

# 그래프 저장
plt.savefig('/content/drive/MyDrive/opensource/sine_wave_plot.png')  # 그래프를 파일로 저장

# 그래프 화면에 표시
plt.show()
```
